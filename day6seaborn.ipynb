```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = px.data.tips()
df.head()
```

```python
df.isnull().sum()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(30,15))
sns.displot(x = df['total_bill'],kde = True)
```

```python
sns.regplot(x=df['total_bill'], y=df['tip'])
```

```python
sns.jointplot(x=df['total_bill'], y=df['tip'])
```

```python
sns.scatterplot(x=df['total_bill'], y=df['tip'])
```

```python
sns.pairplot(df)
```

```python
sns.countplot(x=df['day'])
```

```python
plt.figure(figsize=(12,8))
ax = sns.countplot(x='day',data=df,color='black')

#add counts on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center',            va='baseline', fontsize=10, color='black', xytext=(0,5), textcoords='offset points')

plt.title('Day Distribution')
plt.xticks(rotation=90)
plt.legend(loc=[1,0])
plt.show()
```

```python
plt.figure(figsize=(12,8))
ax = sns.barplot(x='day',y='total_bill',data=df,palette='dark')

#add counts on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center',va='baseline', fontsize=10, color='black', xytext=(0,5), textcoords='offset points')

plt.title('Day Distribution')
plt.xticks(rotation=90)
plt.legend(loc=[1,0])
plt.show()
```

```python
plt.figure(figsize=(12,8))
ax = sns.barplot(x='day',y='total_bill',data=df,palette='dark', estimator=np.max)

#add counts on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center',va='baseline', fontsize=10, color='black', xytext=(0,5), textcoords='offset points')

plt.title('Day Distribution')
plt.xticks(rotation=90)
plt.legend(loc=[1,0])
plt.show()
```

```python
#hue: are used for extra comparison
plt.figure(figsize=(12,8))
ax = sns.barplot(x='day',y='total_bill',data=df,palette='dark', estimator=np.max, hue='sex')

#add counts on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center',va='baseline', fontsize=10, color='black', xytext=(0,5), textcoords='offset points')

plt.title('Day Distribution')
plt.xticks(rotation=90)
plt.legend(loc=[1,0])
plt.show()
```

```python
plt.figure(figsize=(12,8))
ax = sns.countplot(x='day',data=df, palette='dark',hue='sex')

#add counts on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center',            va='baseline', fontsize=10, color='black', xytext=(0,5), textcoords='offset points')

plt.title('Day Distribution')
plt.xticks(rotation=90)
plt.legend(loc=[1,0])
plt.show()
```
