```python
#handling missing values
df.isnull().sum()
```

```python
#forward fill
df2 = df.copy()
df2['Units'].fillna(method='ffill',inplace=True)
df2.tail()
```

```python
#backward fill
df2 = df.copy()
df2['Units'].fillna(method='bfill',inplace=True)
df2.tail()
```

```python
#statistical way
df3 = df.copy()
df3['Units'].fillna(np.mean(df3['Units']), inplace=True)
df3.tail()
```

```python
#replacement method
df4 = df.copy()
df4['Units'].replace({np.NaN:19}, inplace=True)
df4.tail()
```

```python
#drop method
df.dropna(inplace=True)
df.tail()
```

```python
#converting a date column from object(string) to datetime
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
#how to sort values
df.sort_values(by = 'Date', ignore_index=True, ascending=False,inplace=True)
df.head()
```

```python
#apply method
df.apply('mean', numeric_only = True)
```

```python
df.apply('max')
```

```python
#using groupby with the apply method
df.groupby('Region').apply('mean', numeric_only = True)
```

```python
df.groupby(['Region','Type']).apply('mean', numeric_only=True)
```

```python
#pivot table
pvt = pd.pivot_table(df,index= 'Region', values= 'Sales', columns= 'Type')
pvt
```

```python
pvt = pd.pivot_table(df,index= 'Region', values= 'Sales', columns= 'Type', aggfunc= 'max')
pvt
```

```python
pvt = pd.pivot_table(df,index= 'Region', values= 'Sales', columns= 'Type', aggfunc= np.min)
pvt
```
