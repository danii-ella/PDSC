```python
#unsupervised machine learning: to group data based on similar features
#only 2 columns can be used
#we don't have independent and dependent variable
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import plotly.express as px
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
X = df.iloc[:,[3,4]].values
```

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit(X)
    wcss.append(km.inertia_)
wcss
```

```python
plt.style.use('ggplot')
plt.plot(range(1,11),wcss)
plt.xlabel('RANGE')
plt.ylabel('WCSS')
plt.title('ELBOW METHOD')
plt.show()
```

```python
ymeans = KMeans(n_clusters=5)
ypred = ymeans.fit_predict(X)
ypred
```

```python
plt.scatter(X[ypred==0,0], X[ypred==0,1], c='pink', s=200)
plt.scatter(X[ypred==1,0], X[ypred==1,1], c='red', s=200)
plt.scatter(X[ypred==2,0], X[ypred==2,1], c='green', s=200)
plt.scatter(X[ypred==3,0], X[ypred==3,1], c='blue', s=200)
plt.scatter(X[ypred==4,0], X[ypred==4,1], c='cyan', s=200)
plt.scatter(ymeans.cluster_centers_[:,0], ymeans.cluster_centers_[:,1], c='black', s=100, label='CLUSTER POINT')
plt.xlabel('ANNUAL INCOME')
plt.ylabel('SPENDING SCORE')
plt.title('GROUPING OF CUSTOMERS BASED ON THEIR ANNUAL INCOME AND SPENDING SCORE', fontsize=15, color='black')
plt.legend(loc=[1,0])
plt.show()
```
