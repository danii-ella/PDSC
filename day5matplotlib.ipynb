```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gdp = pd.read_csv('RealGDP20122023.csv')
gdp.head()
```

```python
gdp.reset_index(inplace=True)
gdp.head()
```

```python
gdp.rename({'index':'tyear', 'tyear':'Period', 'Period':'Agriculture', 'Agriculture':'CropProduction', 'CropProduction':'Livestock','Livestock':'Forestry', 'Forestry':'Fishing', 'Fishing':'Industry', 'Industry':'MiningAndQuarrying', 'MiningAndQuarrying':'CrudePetroleumAndNaturalGas','CrudePetroleumAndNaturalGas':'CoalMining', 'CoalMining':'MetalOres', 'MetalOres':'QuarryingAndOtherMinerals','QuarryingAndOtherMinerals':'Manufacturing', 'Manufacturing':'OilRefining', 'OilRefining':'Cement', 'Cement':'FoodBeverageAndTobacco', 'FoodBeverageAndTobacco':'TextileApparelAndFootwear','TextileApparelAndFootwear':'WoodAndWoodProducts','WoodAndWoodProducts':'PulpPaperAndPaperProducrs', 'PulpPaperAndPaperProducts':'ChemicalAndPharmaceuticalProducts' ,'ChemicalAndPharmaceuticalProducts':'NonMetallicProducts', 'NonMetallicProducts':'PlasticAndRubberProducts' ,'PlasticAndRubberProducts':'ElectricalAndElectronics', 'ElectricalAndElectronics':'BasicMetalIronAndSteel',
'BasicMetalIronAndSteel':'MotorVehicleAndAssembly', 'MotorVehiclesAndAssembly':'OtherManufacturing',
'OtherManufacturing':'ElectricGasSteamAndAirCon', 'ElectricityGasSteamAndAirCon':'WaterSupplySewageWaste',
'WaterSupplySewageWaste':'Construction', 'Construction':'Services', 'Services':'Trade', 'Trade':'AccomodationAndFoodServices', 'AccommodationAndFoodServices':'TransportationAndStorage', 'TransportationAndStorage':'RoadTransport', 'RoadTransport':'RailTransportAndPipelines', 'RailTransportAndPipelines':'WaterTransport', 'WaterTransport':'AirTransport', 'AirTransport':'TransportServices', 'TransportServices':'PostAndCourierServices', 'PostAndCourierServices':'InformationAndCommunication', 'InformationAndCommunication':'Publishing', 'Publishing':'MotionPicturesSoundRecordingAndMusicProduction', 'MotionPicturesSoundRecordingAndMusicProduction':'Broadcasting', 'Broadcasting':'ArtsEntertainmentAndRecreation', 'ArtsEntertainmentAndRecreation':'FinanceAndInsurance', 'FinanceAndInsurance':'FinancialInstitutions', 'FinancialInstitutions':'Insurance', 'Insurance':'RealEstate', 'RealEstate':'ProffessionalScientificAndTechnicalServices', 'ProfessionalScientificAndTechnicalServices':'AdministrativeAndSupportServicesBusinessServices',
'AdministrativeAndSupportServicesBusinessServices':'PublicAdministration', 'PublicAdministration':'Education', 'Education':'HumanHealthAndSocialServices', 'HumanHealthAndSocialServices':'OtherServices', 'OtherServices':'GDP at 2010 Constant Basic Prices', 'GDP at 2010 Constant Basic Prices':'NetTaxesOnProducts',
'NetTaxesOnProducts':'GDP at 2010 Constant Market Prices', 'GDP at 2010 Constant Market Prices':'None'}, axis=1, inplace=True)
```

```python
gdp.drop('None',axis=1, inplace=True)
gdp.head()
```

```python
plt.plot(gdp['tyear'], gdp['Agriculture'])
```

```python
gdp2 = gdp[gdp['Period'] == 'Annual']
plt.style.use('ggplot')
plt.plot(gdp2['tyear'], gdp2['Agriculture'], color= 'black')
plt.xlabel('YEAR', color= 'black', fontsize= 20)
plt.ylabel('GDP', color= 'black', fontsize= 20)
plt.title('TREND OF AGRICULTURAL GDP OVER TIME', color ='black', fontsize = 25)
plt.show()
```

```python
fig,axes = plt.subplots(3,3,figsize = (18,15))
axes[0,0].plot(gdp2['tyear'], gdp2['Agriculture'],color='magenta')
axes[0,0].set_xlabel('YEAR',color='green',fontsize = 10)
axes[0,0].set_ylabel('GDP',color='green',fontsize = 10)
axes[0,0].set_title('TREND OF AGRICULTURAL GDP OF TIME',color='green',fontsize=10)

axes[0,1].plot(gdp2['tyear'], gdp2['Agriculture'],color='magenta')
axes[0,1].set_xlabel('YEAR',color='green',fontsize = 10)
axes[0,1].set_ylabel('GDP',color='green',fontsize = 10)
axes[0,1].set_title('COMPARISON OF AGRICULTURAL GDP OF YEARLY',color='green',fontsize=10)

axes[0,2].plot(gdp2['tyear'], gdp2['Agriculture'],color='magenta')
axes[0,2].set_xlabel('YEAR',color='green',fontsize = 10)
axes[0,2].set_ylabel('GDP',color='green',fontsize = 10)
axes[0,2].set_title('TREND OF AGRICULTURAL GDP OF TIME',color='green',fontsize=10)

axes[1,0].step(gdp2['tyear'], gdp2['Education'],color='magenta')
axes[1,0].set_xlabel('YEAR',color='green',fontsize = 10)
axes[1,0].set_ylabel('GDP',color='green',fontsize = 10)
axes[1,0].set_title('TREND OF EDUCATIONAL GDP OF TIME',color='green',fontsize=10)

axes[1,1].step(gdp2['tyear'], gdp2['ProffessionalScientificAndTechnicalServices'],color='magenta')
axes[1,1].set_xlabel('YEAR',color='green',fontsize = 10)
axes[1,1].set_ylabel('GDP',color='green',fontsize = 10)
axes[1,1].set_title('TREND OF PSTS GDP OF TIME',color='green',fontsize=10)

axes[1,2].step(gdp2['tyear'], gdp2['CrudePetroleumAndNaturalGas'],color='magenta')
axes[1,2].set_xlabel('YEAR',color='green',fontsize = 10)
axes[1,2].set_ylabel('GDP',color='green',fontsize = 10)
axes[1,2].set_title('TREND OF CRUDEOIL GDP OF TIME',color='green',fontsize=10)

axes[2,0].bar(gdp2['tyear'], gdp2['WoodAndWoodProducts'],color='magenta')
axes[2,0].set_xlabel('YEAR',color='green',fontsize = 10)
axes[2,0].set_ylabel('GDP',color='green',fontsize = 10)
axes[2,0].set_title('COMPARISON OF WOOD AND WOOD GDP OF YEARLY',color='green',fontsize=10)

axes[2,1].bar(gdp2['tyear'], gdp2['AirTransport'],color='magenta')
axes[2,1].set_xlabel('YEAR',color='green',fontsize = 10)
axes[2,1].set_ylabel('GDP',color='green',fontsize = 10)
axes[2,1].set_title('COMPARISON OF AIRTRANSPORT GDP OF YEARLY',color='green',fontsize=10)

axes[2,2].bar(gdp2['tyear'], gdp2['FinancialInstitutions'],color='magenta')
axes[2,2].set_xlabel('YEAR',color='green',fontsize = 10)
axes[2,2].set_ylabel('GDP',color='green',fontsize = 10)
axes[2,2].set_title('COMPARISON OF FINANCIAL INSTITUTION GDP OF YEARLY',color='green',fontsize=10)
```

```python
gdp3 = gdp2[['Agriculture', 'CrudePetroleumAndNaturalGas', 'Education', 'RealEstate']]
gdp3.loc['Total'] = (sum(gdp['Agriculture']), sum(gdp['CrudePetroleumAndNaturalGas']), sum(gdp['Education']), sum(gdp['RealEstate']))
gdp3.tail()
```
