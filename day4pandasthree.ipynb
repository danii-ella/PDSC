```python


df_rate = pd.read_csv('exchange27022024.csv')
df_rate.head()
```

```python
#merge, concat, save to csv, save to excel, save to multiple excel sheet
#merge works just like it would in mysql, it is used to join only 2 tables
USD_table = df_rate[df_rate['Currency'] == 'USD']
EURO_table = df_rate[df_rate['Currency'] == 'EURO']
POUNDS_table = df_rate[df_rate['Currency'] == 'POUNDS']
NAIRA_table = df_rate[df_rate['Currency'] == 'NAIRA']
```

```python
#outer: used for merging tables that have the same heading
df_merge = USD_table.merge(EURO_table, how= 'outer')
df_merge
```

```python
#concatenation(used to join more than 2 tables)
alldf = pd.concat([USD_table, EURO_table, POUNDS_table, NAIRA_table], axis=0)
alldf
```

```python
df_rate
```

```python
#how to save to csv
alldf.to_csv('Rate.csv', index=False)
```

```python
#how to save to excel
alldf.to_excel('Rate.xlsx', index=False, sheet_name= 'EXCHANGERATE')
```

```python
with pd.ExcelWriter('ALLCURR_RATE.xlsx', engine= 'xlsxwriter') as writer:
    USD_table.to_excel(writer, index= False, sheet_name= 'USD', engine='xlsxwriter')
    EURO_table.to_excel(writer, index= False, sheet_name= 'EURO', engine='xlsxwriter')
    POUNDS_table.to_excel(writer, index= False, sheet_name= 'POUNDS', engine='xlsxwriter')
    NAIRA_table.to_excel(writer, index= False, sheet_name= 'NAIRA', engine='xlsxwriter')
```
